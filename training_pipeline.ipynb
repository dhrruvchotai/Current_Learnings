```python
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
df.head()
```

```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
```

```python
df["diagnosis"].value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=["diagnosis"]),df["diagnosis"],test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train = torch.from_numpy(X_train)
X_test = torch.from_numpy(X_test)
y_train = torch.from_numpy(y_train)
y_test = torch.from_numpy(y_test)
```

```python
class Model:
    def __init__(self,X):
        self.weights = torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
        self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)

    def forward(self,X):
        z = torch.matmul(X,self.weights)+self.bias
        y_pred = torch.sigmoid(z)
        return y_pred
    
    def binary_cross_entropy(self,y_train,y_pred):
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred,min=epsilon,max=1-epsilon)
        loss = -(y_train*torch.log(y_pred)+(1-y_train)*torch.log(y_pred)).mean()
        return loss
```

```python
learning_rate = 0.1
epochs = 25
```

```python
model = Model(X_train)
```

```python
model.weights
```

```python
model.bias
```

```python
X_train.dtype
```

```python
for epoch in range(epochs):
    y_pred = model.forward(X_train)
    loss = model.binary_cross_entropy(y_train,y_pred)
    print(f"Epoch : {epoch+1} Loss : {loss.item()}")
    loss.backward()

    with torch.no_grad():
        model.weights -= learning_rate*model.weights.grad
        model.bias -= learning_rate*model.bias.grad

    model.weights.grad.zero_()
    model.bias.grad.zero_()
    
```

```python
with torch.no_grad():
    y_pred = model.forward(X_test)
    y_pred = (y_pred>0.5).float()
    accuracy = (y_pred == y_test).float().mean()
    print(f"Accuracy : {accuracy}")
```
